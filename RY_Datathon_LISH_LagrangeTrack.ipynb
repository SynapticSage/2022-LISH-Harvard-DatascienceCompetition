# **Data Challenge Overview -- Pump it Up: Data Mining the Water Table**

Access to potable water is vital to the health and wellbeing of communities around the world. A smart understanding of which waterpoints will fail can improve maintenance operations and ensure that clean, potable water is available everywhere. ### Feature Engineering

Ready both train and testFeatures(for later)

```python
#Combine variables
usefulvars = ["quantity_group", "quality_group", "public_meeting", "extraction_type_class", "management_group", "basin", "payment", "source", "permit", "region", "scheme_management", "lga"]

def normalize(x):
  x = (x - np.min(x, axis=None))/(np.max(x, axis=None)-np.min(x, axis=None))
  x = x-np.median(x)
  return x

def ready_variables(train:pd.DataFrame, transform:dict=None):
  train = train.copy()
  if transform is None:
    transform = dict()
  # -------- Combine highly redundancy features ---------
  train.source = train.source + "_" + train.source_type + "_" + train.source_class
  train.drop(columns=["source_type","source_class"], inplace=True)
  train.extraction = train.extraction_type + "_" + train.extraction_type_class + "_" + train.extraction_type_group
  train.drop(columns=["extraction_type", "extraction_type_class", "extraction_type_group"])
  # -------- Drop out less useful features (from up top plots) ---------
  less_useful_variables = list(set(train.select_dtypes(object)) - set(usefulvars))
  train.drop(columns=[*less_useful_variables, "id"], inplace=True)

  # -------- Hot encode categoricals for modeling purposes ---------
  # List out the categoricals : we need to hot encoode for the model of interest
  categoricals = train.select_dtypes(object)
  # and intiate hot encoding
  from sklearn.preprocessing import OneHotEncoder
  encoder = OneHotEncoder(handle_unknown="ignore")
  train_preprocess = {}
  categorical_ints = ["region_code", "district_code"]
  for name in [*categoricals, *categorical_ints]:
    if name in transform:
      pass
    else:
      transform[name] = encoder.fit(train[name].values.reshape(-1,1)).transform
    train_preprocess[name] = transform[name](train[name].values.reshape(-1,1)).toarray().astype('int8')
  # some numericals are hot encoded and need to be categorical
  numericals = set(train.select_dtypes([float, int]).columns) - set(["status_group", *categorical_ints])
  for name in numericals:
    train_preprocess[name] = np.atleast_2d(train[name].values).T
  {x:y.shape for x,y in train_preprocess.items()}
  # let's make the ranges equal, so that variablees aren't important by virtue of size
  train_preprocess = {key:normalize(item) for key,item in train_preprocess.items()}
  X = np.hstack([item for key,item in train_preprocess.items()])
  if "status_group" in train:
    y = train.status_group.astype('float').values
    return X, y, train_preprocess, transform
  else:
    return X

overall_set = pd.concat([train, testFeatures], keys=['train','test']).reset_index()
train_idx = overall_set.level_0 == "train"
test_idx = overall_set.level_0 == "test"
(Xoverall, yoverall, preprocess, transform) = ready_variables(overall_set)
X, y = Xoverall[train_idx ,:], yoverall[train_idx]
Xtest = Xoverall[test_idx,:]
```
